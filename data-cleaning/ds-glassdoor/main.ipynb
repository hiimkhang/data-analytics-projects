```python
data[['Min Salary', 'Max Salary']] = data['Salary Estimate'].apply(
    lambda x: pd.Series(re.findall(pattern='\d+', string=x))
)
```

```python
data.drop(['index', 'Salary Estimate', 'Founded', 'Competitors'], axis=1, inplace=True)
```

```python
data['Company Name'] = data['Company Name'].apply(
    lambda x: pd.Series(str(x).split(sep='\n')[0])
)
```

```python
senior_keywords = ["Sr.", "Sr", "Senior"]
junior_keywords = ["Jr.", "Jr", "Junior"]
```

```python
def extractExperienceLevel(record: str):
    extracted = ''

    if any(keyword in record for keyword in senior_keywords):
        extracted = "Senior"
    elif any(keyword in record for keyword in junior_keywords):
        extracted = "Junior"
    else:
        extracted = "Not specified"
        
    for keyword in senior_keywords + junior_keywords:
        record = record.replace(keyword, "")
        
    return extracted
```

```python
data['Experience Level'] = data['Job Title'].apply(extractExperienceLevel)

data['Job Title'] = data['Job Title'].apply(
    lambda x: pd.Series(re.sub(pattern=r"(?:Sr\.?|\(Sr\.?\)|Senior|Junior|Jr\.?) ", string=x, repl=""))
)

data['Job Title'] = data['Job Title'].apply(
    lambda x: pd.Series(re.sub(pattern=r"(, (?:Sr\.?|\(Sr\.?\)|Senior))", string=x, repl=""))
)
```

```python
data['State Code'] = data['Location'].apply(
    lambda x: str(x).split(sep=',')[-1].strip())
data['State Code'].replace(['United States', 'New Jersey', 'Texas', 'California', 'Utah', 'Remote'], 
                           ['US', 'NJ', 'TX', 'CA', 'UT', np.nan])
```

```python
data = data[['Job Title', 'Job Description', 'Experience Level', 'Company Name', 'Min Salary', 'Max Salary', 'Rating',
             'Location', 'Headquarters', 'State Code', 'Size', 'Type of ownership', 'Industry', 'Sector', 'Revenue']]
```
